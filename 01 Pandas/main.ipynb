```python
import pandas as pd
df = pd.read_csv('winequality-red.csv', sep=',')
df.head()
```

```python
results = [
    df.mean(),
    df.median(),
    df.std(),
    df.var(),
    df.mode().iloc[0],
    df.max() - df.min(),
]

index_keys = ['mean', 'median', 'standard deviation', 'variation', 'mode', 'range of values']
for i, key in enumerate(index_keys):
    results[i].name = key

results_df = pd.concat(results, axis=1)

results_df
```
